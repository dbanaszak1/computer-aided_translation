### Ćwiczenie 1: Napisz funkcję do pobierania nazw towarów z serwisu Ceneo.pl. Typ towaru, np. telewizor, pralka, laptop jest parametrem funkcji. Wystarczy pobierać dane z pierwszej strony wyników wyszukiwania.

```python
import json

def get_names(article_type: str):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36',
    }

    url = f"https://www.ceneo.pl/{article_type}"

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.content, 'html.parser')

        names = []

        scripts = soup.find_all('script', type='application/ld+json')

        for script in scripts:
            if not script.string:
                continue
            try:
                data = json.loads(script.string)

                if isinstance(data, dict) and data.get('@type') == 'ItemList':
                    items = data.get('itemListElement', [])
                    for entry in items:
                        product_name = entry.get('item', {}).get('name')
                        if product_name:
                            names.append(product_name.strip())
            except json.JSONDecodeError:
                continue

        if not names:
            elements = soup.select('strong.cat-product__name, a.grid-item__name, [data-name]')
            for el in elements:
                name = el.get('data-name') or el.get_text()
                if name:
                    names.append(name.strip())

        return list(dict.fromkeys(names))

    except Exception as e:
        print(f"Błąd: {e}")
        return []

products = get_names("Telewizory")
for name in products:
    print(name)
```

W ten sposób pobieramy dane z jednej strony. Nic jednak nie stoi nam na przeszkodzie, aby zasymulować przełączanie stron.

### Ćwiczenie 3: Napisz program do pobrania tekstu ze strony Wydziału Matematyki i Informatyki. Pobierz cały tekst ze strony głównej a następnie wyszukaj na tej stronie wszystkich linków wewnętrznych i pobierz tekst ze stron wskazywanych przez te linki. Nie zagłębiaj się już dalej.


```python
import time
from urllib.parse import urlparse, urljoin

def get_clean_text(soup):
    for element in soup(["script", "style", "noscript", "meta"]):
        element.decompose()
    text = soup.get_text(separator=" ", strip=True)
    return re.sub(r"\s+", " ", text)

def scrape_wmi():
    base_url = "https://wmi.amu.edu.pl/"
    domain = urlparse(base_url).netloc
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
    }
    session = requests.Session()
    session.headers.update(headers)

    print(f"Downloading: {base_url}")
    response = session.get(base_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    internal_links = set()

    for a_tag in soup.find_all('a', href=True):
        href = a_tag['href']

        full_url = urljoin(base_url, href)

        parsed_url = urlparse(full_url)

        if parsed_url.netloc == domain:
            clean_url = f"{parsed_url.scheme}://{parsed_url.netloc}{parsed_url.path}"

            if not clean_url.lower().endswith(('.pdf', '.jpg', '.png', '.doc', '.zip')):
                internal_links.add(clean_url)

    main_page_text = get_clean_text(soup)
    print(f"{main_page_text[:150]}...\n")

    print(f"Found {len(internal_links)} links, downloading...\n")

    all_scraped_data = {}
    all_scraped_data[base_url] = main_page_text

    for i, link in enumerate(list(internal_links), 1):
        print(f"[{i}/{len(internal_links)}] downloading: {link}")
        try:
            res = session.get(link, timeout=10)

            if 'text/html' in res.headers.get('Content-Type', ''):
                page_soup = BeautifulSoup(res.content, 'html.parser')
                all_scraped_data[link] = get_clean_text(page_soup)

        except Exception as e:
            print(f"Error {link}: {e}")

        time.sleep(0.5)

    return all_scraped_data

dane = scrape_wmi()
```

```python
def get_common_english_words(limit=3000):
    url = "https://raw.githubusercontent.com/first20hours/google-10000-english/master/google-10000-english-no-swears.txt"
    try:
        response = requests.get(url)
        response.raise_for_status()
        all_words = response.text.splitlines()
        return all_words[:limit]
    except Exception as e:
        return ["test", "house", "water"]

english_base = get_common_english_words(1000)
```
