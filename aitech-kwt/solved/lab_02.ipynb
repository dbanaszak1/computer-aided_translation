### Ćwiczenie 1: Wykonaj analizę funkcjonalności pamięci tłumaczeń w programach SDL Trados Studio 2021 oraz Kilgray memoQ. Dla obu programów wypisz funkcje, które są związane z TM oraz zaznacz, które funkcje są wspólne dla obu programów oraz których funkcji Tradosa brakuje w memoQ oraz odwrotnie.

Odpowiedź:

Wspólne funkcje pamięci tłumaczeń (TM)
* **Zarządzanie formatami wymiany:** Import i eksport plików TMX (Translation Memory eXchange), co pozwala na migrację pamięci między programami.
* **Poziomy dopasowań (Matches):** Obsługa dopasowań stuprocentowych (100%), rozmytych (Fuzzy matches, zazwyczaj od 50% do 99%) oraz dopasowań kontekstowych (Context Matches w Tradosie / 101% w memoQ).
* **Wyszukiwanie konkordancyjne (Concordance Search):** Możliwość ręcznego przeszukiwania TM pod kątem konkretnych słów, fraz lub wyrażeń w tekście źródłowym i docelowym.
* **Naprawa dopasowań rozmytych (Fuzzy Match Repair):** Automatyczna korekta dopasowań rozmytych poprzez podmianę słów na podstawie baz terminologicznych (w Tradosie: *upLIFT Match Repair*, w memoQ: *Match Patching*).
* **Kary do dopasowań (Penalties):** Konfigurowalne kary procentowe za różnice w tagach, formatowaniu czy pochodzeniu pamięci.
* **Edycja i konserwacja TM:** Filtrowanie segmentów, globalne wyszukiwanie/zamiana, oznaczanie duplikatów i usuwanie błędnych wpisów.
* **Aktualizacja w czasie rzeczywistym:** Zapisywanie zatwierdzonych segmentów od razu do głównej pamięci tłumaczeń.

Unikalne funkcje Trados Studio 2021
* **Technologia AnyTM jako standardowy silnik:** Pozwala na podpięcie do projektu pamięci tłumaczeniowej o zupełnie innych językach (np. użycie pamięci EN-ES w projekcie EN-DE). W memoQ system rygorystycznie pilnuje par językowych.
* **Modele upLIFT Fragment Recall:** Technologia wyrównująca słowa i frazy wewnątrz samej pamięci, tworząca precyzyjny słownik podsegmentów na poziomie silnika TM (nawet bez bazy terminologicznej).
* **Zaawansowany edytor Alignmentu:** Dedykowany, rozbudowany widok do precyzyjnego łączenia i dzielenia segmentów z dokumentów "legacy" w celu stworzenia pliku `.sdltm`.
* **Ekosystem RWS AppStore:** Głęboka modyfikacja pracy z TM dzięki setkom wtyczek zewnętrznych (np. aplikacje do anonimizacji TM).

Unikalne funkcje memoQ
* **LiveDocs (Aktywne Korpusy):** Możliwość wrzucenia dwóch przetłumaczonych wcześniej plików bezpośrednio do bazy. Program wyrównuje je "w locie" i podpowiada tłumaczenia jak ze standardowej TM. Trados wymaga tu pełnego procesu Alignment.
* **Natywna dwukierunkowość wszystkich TM:** Każda pamięć w memoQ jest z definicji odwracalna (np. TM z PL-EN może być użyta w projekcie EN-PL). W Tradosie technicznie TM są jednokierunkowe (wymaga to użycia AnyTM).
* **Metadane na poziomie segmentu:** Łatwiejsze śledzenie ról użytkowników (np. Translator, Reviewer 1) przypisanych do konkretnego segmentu w TM, co pozwala na zaawansowane filtrowanie.
* **Kary oparte na rolach (Role-based penalties):** Możliwość zdefiniowania kar do dopasowań w zależności od tego, kto zapisał dany segment do TM (np. -5% kary za segment od początkującego tłumacza).


Jedną z funkcji dostępnych we wszystkich większych programach do wspomagania tłumaczenia jest znajdowanie bardzo pewnych dopasowań w pamięci tłumaczeń. Są one zwane **ICE** (In-Context Exact match) lub 101% match. Są to takie dopasowania z pamięci tłumaczeń, dla których nie tylko zdanie źródłowe z TM jest identyczne z tłumaczonym, ale także poprzednie zdanie źródłowe z TM zgadza się z poprzednim zdaniem tłumaczonym oraz następne z TM z następnym tłumaczonym.

### Ćwiczenie 2: Zaimplementuj funkcję ice_lookup, przyjmującą trzy parametry: aktualnie tłumaczone zdanie, poprzednio tłumaczone zdanie, następne zdanie do tłumaczenia. Funkcja powinna zwracać dopasowania typu ICE. Nie pozwól, aby doszło do błędów podczas sprawdzania pierwszego i ostatniego przykładu w pamięci (ze względu na brak odpowiednio poprzedzającego oraz następującego przykładu).

```python
def ice_lookup(sentence, prev_sentence, next_sentence):
    matches = []
    for i in range(len(translation_memory)):
        if translation_memory[i][0] == sentence:
            if i > 0:
                prev_match = (translation_memory[i-1][0] == prev_sentence)
            else:
                prev_match = (prev_sentence is None)

            if i < len(translation_memory) - 1:
                next_match = (translation_memory[i+1][0] == next_sentence)
            else:
                next_match = (next_sentence is None)
                
            if prev_match and next_match:
                matches.append(translation_memory[i][1])
                
    return matches


print(ice_lookup(
    sentence='Drukarka jest wyłączona', 
    prev_sentence='Sprawdź ustawienia sieciowe', 
    next_sentence='Wymagane ponowne uruchomienie komputera'
))

print(ice_lookup(
    sentence='Wciśnij przycisk Enter', 
    prev_sentence=None, 
    next_sentence='Sprawdź ustawienia sieciowe'
)) 

print(ice_lookup(
    sentence='Wymagane ponowne uruchomienie komputera', 
    prev_sentence="Drukarka jest wyłączona", 
    next_sentence=None
)) 
```

Inną powszechnie stosowaną techniką przeszukiwania pamięci tłumaczeń jest tzw. **fuzzy matching**. Technika ta polega na wyszukiwaniu zdań z pamięci, które są tylko podobne do zdania tłumaczonego. Na poprzednich zajęciach wykonywaliśmy funkcję tm_lookup, która pozwalała na różnicę jednego słowa.

Zazwyczaj jednak funkcje fuzzy match posiadają znacznie szersze możliwości. Ich działanie opiera się na zdefiniowaniu funkcji $d$ dystansu pomiędzy zdaniami $x$ i $y$. Matematycznie, funkcja dystansu posiada następujące właściwości:
1. $\forall_{x,y} d(x,y)\geqslant 0$
2. $\forall_{x,y} d(x,y)=0 \Leftrightarrow x=y$
3. $\forall_{x,y} d(x,y)=d(y,x)$
4. $\forall_{x,y,z} d(x,y) + d(y,z)\geqslant d(x,z)$

Rozważmy następującą funkcję:

```python
def sentence_distance(x,y):
    return abs(len(y) - len(x))
```

### Ćwiczenie 4: Czy to jest poprawna funkcja dystansu? Które warunki spełnia?

Tak,funkcja spełnia wszystkie wymagane warunki matematyczne.

Wprowadźmy jednak inną funkcję dystansu - dystans Levenshteina. Dystans Levenshteina pomiędzy dwoma łańcuchami znaków definiuje się jako minimalną liczbę operacji edycyjnych, które są potrzebne do przekształcenia jednego łańcucha znaków w drugi. Wyróżniamy trzy operacje edycyjne:
* dodanie znaku
* usunięcie znaku
* zamiana znaku na inny

### Ćwiczenie 5: Czy dystans Levenshteina jest poprawną funkcją dystansu? Uzasadnij krótko swoją odpowiedź sprawdzając każdy z warunków.

Dystans Levenshteina jest metryką, bo spełnia cztery warunki:

- Nieujemność - liczba operacji edycyjnych jest zawsze ≥ 0.
- Identyczność - dystans wynosi 0 tylko dla identycznych napisów.
- Symetria - przekształcenie \(x \to y\) kosztuje tyle samo co \(y \to x\).
- Nierówność trójkąta - przejście przez pośredni napis nie skraca drogi, więc \(d(x,y) + d(y,z) \ge d(x,z)\).

```python
from Levenshtein import distance as levenshtein_distance

levenshtein_distance("kotek", "kotki")
```

Funkcja ta daje nam możliwość zdefiniowania podobieństwa pomiędzy zdaniami:

```python
def levenshtein_similarity(x,y):
    return 1 - levenshtein_distance(x,y) / max(len(x), len(y))
```

### Ćwiczenie 6: Napisz funkcję fuzzy_lookup, która wyszuka w pamięci tłumaczeń wszystkie zdania, których podobieństwo Levenshteina do zdania wyszukiwanego jest większe lub równe od ustalonego progu.

```python
def fuzzy_lookup(sentence, threshold):
    results = []
    for source, target in translation_memory:
        similarity = levenshtein_similarity(sentence, source)
        if similarity >= threshold:
            results.append((source, target, similarity))

    results.sort(key=lambda x: x[2], reverse=True)
    
    return results

print(translation_memory)
print(fuzzy_lookup('Wciśnij klawisz Enter', 0.5))
print(fuzzy_lookup('Wciśnij klawisz Enter', 0.9))
```
